```python
import pandas as pd
from sklearn import preprocessing
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("C:/Users/user2/Desktop/spam1.csv",encoding="ISO-8859-1")
df
```

```python
df.info()
```

```python
df.shape
```

```python
label=preprocessing.LabelEncoder()
label1=label.fit_transform(df["v1"])
label1
```

```python
df["v1"]=label1
df
```

```python
df.describe()
```

```python

cvt=CountVectorizer()
x=cvt.fit_transform(df["v2"])
arry=x.toarray()
arry
```

```python
df=pd.DataFrame(data=arry,columns=cvt.get_feature_names())
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,label1)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
clf.fit(x_train,y_train)
predict=clf.predict(x_test)
predict
```
